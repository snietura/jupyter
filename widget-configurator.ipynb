```python
widget_repo = {'Suwak liczb całkowitych': 
                    {'desc': 'Suwak do określania liczby całkowitej z wybranego zakresu',
                     'cons': 'IntSlider({})',
                     'cat' : 'numeric',
                     'props': {'value': 
                                    {'type': 'int',
                                     'desc': 'wybrana wartość suwaka',
                                     'def' : 0
                                    },
                               'min': 
                                    {'type': 'int',
                                     'desc': 'wartość minimalna suwaka',
                                     'def' : 0
                                    },
                               'max':
                                    {'type': 'int',
                                     'desc': 'wartość maksymalna suwaka',
                                     'def' : 100
                                    },
                               'step': 
                                    {'type': 'int',
                                     'desc': 'skok suwaka',
                                     'def' : 1
                                    },
                               'description': 
                                    {'type': 'string',
                                     'desc': 'opis/etykieta wyświetlany przy suwaku',
                                     'def' : ''
                                    },
                               'description_tooltip':
                                    {'type': 'string',
                                     'desc': 'opis wyświetlany przy najechaniu kursorem na suwak',
                                     'def' : ''
                                    },
                               'disabled': 
                                    {'type': 'bool',
                                     'desc': 'czy suwak jest wyłączony/nieaktywny',
                                     'def' : 0
                                    },
                               'continuous_update': 
                                    {'type': 'bool',
                                     'desc': 'czy wartość zmienia się podczas przeciągania myszą lub dopiero po puszczeniu',
                                     'def' : 1
                                    },
                               'orientation': 
                                    {'type': 'horizontal|vertical',
                                     'desc': 'orientracja pozioma lub pionowa',
                                     'def' : 'horizontal'
                                    },
                               'readout': 
                                    {'type': 'bool',
                                     'desc': 'czy wyświetlać aktualną wartość w postaci liczbowej',
                                     'def' : 1
                                    },
                               'readout_format': 
                                    {'type': 'string',
                                     'desc': 'string formatujący wartość liczbową',
                                     'def' : 'd'
                                    }
                              }
                    }, 
              'Suwak liczb rzeczywistych':
                    {'desc': 'Suwak do określania liczby rzeczywistej z wybranego zakresu',
                     'cons': 'FloatSlider({})',
                     'cat' : 'numeric',
                     'props': {'value': 
                                    {'type': 'float',
                                     'desc': 'wybrana wartość suwaka',
                                     'def' : 0.0
                                    },
                               'min': 
                                    {'type': 'float',
                                     'desc': 'wartość minimalna suwaka',
                                     'def' : 0.0
                                    },
                               'max':
                                    {'type': 'float',
                                     'desc': 'wartość maksymalna suwaka',
                                     'def' : 100.0
                                    },
                               'step': 
                                    {'type': 'float',
                                     'desc': 'skok suwaka',
                                     'def' : 0.1
                                    },
                               'description': 
                                    {'type': 'string',
                                     'desc': 'opis/etykieta wyświetlany przy suwaku',
                                     'def' : ''
                                    },
                               'description_tooltip':
                                    {'type': 'string',
                                     'desc': 'opis wyświetlany przy najechaniu kursorem na suwak',
                                     'def' : ''
                                    },
                               'disabled': 
                                    {'type': 'bool',
                                     'desc': 'czy suwak jest wyłączony/nieaktywny',
                                     'def' : 0
                                    },
                               'continuous_update': 
                                    {'type': 'bool',
                                     'desc': 'czy wartość zmienia się podczas przeciągania myszą lub dopiero po puszczeniu',
                                     'def' : 1
                                    },
                               'orientation': 
                                    {'type': 'horizontal|vertical',
                                     'desc': 'orientracja pozioma lub pionowa',
                                     'def' : 'horizontal'
                                    },
                               'readout': 
                                    {'type': 'bool',
                                     'desc': 'czy wyświetlać aktualną wartość w postaci liczbowej',
                                     'def' : 1
                                    },
                               'readout_format': 
                                    {'type': 'string',
                                     'desc': 'string formatujący wartość liczbową',
                                     'def' : '.2f'
                                    }
                              }
                    },  
              'Suwak wykładniczy rzeczywistych':
                    {'desc': 'Suwak do określania liczby rzeczywistej w postaci wykładniczej z wybranego zakresu',
                     'cons': 'FloatLogSlider({})',
                     'cat' : 'numeric',
                     'props': {'value': 
                                    {'type': 'float',
                                     'desc': 'wybrana wartość suwaka',
                                     'def' : 0.0
                                    },
                               'base':
                                    {'type': 'float',
                                     'desc': 'podstawa liczby wykładniczej',
                                     'def' : 10.0                                        
                                    },
                               'min': 
                                    {'type': 'float',
                                     'desc': 'wartość minimalna wykładnika',
                                     'def' : 0.0
                                    },
                               'max':
                                    {'type': 'float',
                                     'desc': 'wartość maksymalna wykładnika',
                                     'def' : 10000.0
                                    },
                               'step': 
                                    {'type': 'float',
                                     'desc': 'skok wykładnika liczby',
                                     'def' : 0.1
                                    },
                               'description': 
                                    {'type': 'string',
                                     'desc': 'opis/etykieta wyświetlany przy suwaku',
                                     'def' : ''
                                    },
                               'description_tooltip':
                                    {'type': 'string',
                                     'desc': 'opis wyświetlany przy najechaniu kursorem na suwak',
                                     'def' : ''
                                    },
                               'disabled': 
                                    {'type': 'bool',
                                     'desc': 'czy suwak jest wyłączony/nieaktywny',
                                     'def' : 0
                                    },
                               'continuous_update': 
                                    {'type': 'bool',
                                     'desc': 'czy wartość zmienia się podczas przeciągania myszą lub dopiero po puszczeniu',
                                     'def' : 1
                                    },
                               'orientation': 
                                    {'type': 'horizontal|vertical',
                                     'desc': 'orientracja pozioma lub pionowa',
                                     'def' : 'horizontal'
                                    },
                               'readout': 
                                    {'type': 'bool',
                                     'desc': 'czy wyświetlać aktualną wartość w postaci liczbowej',
                                     'def' : 1
                                    },
                               'readout_format': 
                                    {'type': 'string',
                                     'desc': 'string formatujący wartość liczbową',
                                     'def' : '.2f'
                                    }
                              }
                    },  
              'Zakres liczb całkowitych': 
                    {'desc': 'Suwak do określania zakresu liczb całkowitych - zwraca tuple dwóch liczb',
                     'cons': 'IntRangeSlider({})',
                     'cat' : 'numeric',
                     'props': {'value': 
                                    {'type': 'range',
                                     'desc': 'wybrana wartość suwaka - zwraca tuple 2 liczb',
                                     'def' : '25|75'
                                    },
                               'min': 
                                    {'type': 'int',
                                     'desc': 'wartość minimalna wykładnika',
                                     'def' : 0
                                    },
                               'max':
                                    {'type': 'int',
                                     'desc': 'wartość maksymalna wykładnika',
                                     'def' : 100
                                    },
                               'step': 
                                    {'type': 'int',
                                     'desc': 'skok wykładnika liczby',
                                     'def' : 1
                                    },
                               'description': 
                                    {'type': 'string',
                                     'desc': 'opis/etykieta wyświetlany przy suwaku',
                                     'def' : ''
                                    },
                               'description_tooltip':
                                    {'type': 'string',
                                     'desc': 'opis wyświetlany przy najechaniu kursorem na suwak',
                                     'def' : ''
                                    },
                               'disabled': 
                                    {'type': 'bool',
                                     'desc': 'czy suwak jest wyłączony/nieaktywny',
                                     'def' : 0
                                    },
                               'continuous_update': 
                                    {'type': 'bool',
                                     'desc': 'czy wartość zmienia się podczas przeciągania myszą lub dopiero po puszczeniu',
                                     'def' : 1
                                    },
                               'orientation': 
                                    {'type': 'horizontal|vertical',
                                     'desc': 'orientracja pozioma lub pionowa',
                                     'def' : 'horizontal'
                                    },
                               'readout': 
                                    {'type': 'bool',
                                     'desc': 'czy wyświetlać aktualną wartość w postaci liczbowej',
                                     'def' : 1
                                    },
                               'readout_format': 
                                    {'type': 'string',
                                     'desc': 'string formatujący wartość liczbową',
                                     'def' : '.2f'
                                    }
                              }
                    },  
              'Zakres liczb rzeczywistych': '',
              'Progress (integer)': '',
              'Progress (float)': '',
              'Bounded text (integer)': '',
              'Bounded text (float)': '',
              'Text (integer)': '',
              'Text (float)': '',
              'Toggle button': '',
              'Checkbox': '',
              'Valid (read-only)': '',
              'Dropdown': '',
              'Radio buttons': '',
              'Select': '',
              'Selection slider': '',
              'Selection range slider': '',
              'Toggle buttons': '',
              'Select multiple': '',
              'Text': '',
              'Text area': '',
              'Label': '',
              'HTML': '',
              'HTML Math': '',
              'Image': '',
              'Button': '',
              'Output': '',
              'Play animation': '',
              'Date picker': '',
              'Color picker': '',
              'File upload': '',
              }

widget_select = widgets.Dropdown(options=widget_repo.items(), description= 'Widget')
widget_desc = widgets.Label(value = widget_select.value['desc'])
widget_props = widgets.VBox()
out = widgets.Output(layout={'border': '1px solid black'})
props = {}


def on_widget_select_change(change):
    global widget_desc, widget_repo, widget_props, props, out
    
    props.clear()
    out.clear_output()
    widget_desc.value = change['new']['desc']
    for prop, var in widget_select.value['props'].items():
        if var['type'] == 'int':
            c = widgets.IntText(value = var['def'])
        elif var['type'] == 'float':
            c = widgets.FloatText(value = var['def'])
        elif var['type'] == 'bool':
            c = widgets.Checkbox(value=bool(var['def']))
        elif var['type'] == 'string':
            c = widgets.Text(value = var['def'])
        elif var['type'] == 'range':
            c = widgets.Text(value=var['def'])
        elif '|' in var['type']:
            c = widgets.Dropdown(options=var['type'].split('|'))
            c.value = var['def']
        c.description = prop
        c.description_tooltip = var['desc']
        props[prop] = c
    widget_props.children = tuple(props.values())


def btn_generuj_on_click(b):
    global props, widget_select, widget_repo, out
    s = ''
    for prop, var in widget_select.value['props'].items():
        if props[prop].value != var['def']:
            if (var['type'] == 'string') or ('|' in var['type']):
                s = s + "{0}='{1}', ".format(prop, props[prop].value) 
            elif var['type'] == 'bool':
                s = s + '{0}={1}, '.format(prop, bool(props[prop].value)) 
            elif var['type'] == 'range':
                s = s + '{0}=({1}), '.format(prop, props[prop].value.replace('|',','))
            else:
                s = s + '{0}={1!s}, '.format(prop, props[prop].value)
    wynik = widget_select.value['cons'].format(s[:-2])
    out.clear_output()
    with out:
        print(wynik)
        exec('display(widgets.{})'.format(wynik))

widget_select.observe(on_widget_select_change, names='value')

s = {'new': widget_select.value}
on_widget_select_change(s)
button = widgets.Button(description='Generuj', icon='check')
button.on_click(btn_generuj_on_click)
display(widget_select, widget_desc, widget_props, button, out)
    
```
